### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)

```python
df_sorted_population = df.sort_values(by="population_total", ascending=False)
top_5_cities = df_sorted_population.iloc[0:5][["city","population_total"]]
top_5_cities["population_total"] = top_5_cities.population_total / 1000
top_5_cities
```

```python
plt.figure(figsize=(7, 6))
bars = plt.bar(top_5_cities["city"], top_5_cities["population_total"], color=["blue", "gray", "brown", "olive", "blue"], alpha=0.7)

for index, bar in enumerate(bars):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 100,
        str(top_5_cities.iloc[index]["population_total"]),
        ha="center",
        va="bottom"
    )
    
plt.title("Ciudades Más Pobladas de California", fontsize=16)
plt.xlabel("Ciudades", fontsize=10)
plt.ylabel("Habitantes (miles)", fontsize=10)
plt.ylim(0, 5000)
plt.show()
```

2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)

```python
plt.figure(figsize=(8, 6))
plt.hist(df["area_total_km2"], bins=50, color="steelblue", alpha=0.7, label="Hist. Área")

plt.legend(fontsize=8, loc="upper right")
plt.xlabel("Superficie (km$^2$)", fontsize=14, fontname="Bradley Hand ITC")
plt.ylabel("Número de ciudades", fontsize=14, fontname="Bradley Hand ITC")
plt.show()
```

3. El scatter de área de agua de un ciudad con su poblacion.

```python
max_area_agua = df.loc[df["area_water_km2"] == df["area_water_km2"].max()]
max_poblacion = df.loc[df["population_total"] == df["population_total"].max()]
```

```python
plt.figure(figsize=(7, 6))
plt.scatter(df["area_water_km2"], df["population_total"] / 1000,alpha=0.6)

plt.title("Población por km$^2$ de agua")
plt.xlabel("Área de agua en la ciudad (km$^2$)", fontsize= 10)
plt.ylabel("Población(miles de habitantes)", fontsize= 10)

plt.annotate(
    "Outliers",
    xy=(max_area_agua["area_water_km2"].iloc[0], max_area_agua["population_total"].iloc[0] / 1000),
    xytext=(60, 3500),  # Coordenadas del texto
    arrowprops=dict(facecolor="black", arrowstyle="->"),
    fontsize=10,
    color="black",
)

plt.annotate(
    "Outliers",
    xy=(max_poblacion["area_water_km2"].iloc[0], max_poblacion["population_total"].iloc[0] / 1000),
    xytext=(60, 3500),  # Coordenadas del texto
    arrowprops=dict(facecolor="black", arrowstyle="->"),
    fontsize=10,
    color="black",
)


plt.show()
```

4. El boxplot de la altura de las ciudades.

```python
plt.figure(figsize=(7, 6))
plt.boxplot(df["elevation_m"], vert= False, patch_artist=True, boxprops=dict(facecolor="red", color="black"), medianprops=dict(color="black"))

plt.title("Diagrama de caja de la altitud de las ciudades de California", fontsize= 10)
plt.xlabel("Altitud (m)")
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# --------------- Primer gráfico ---------------

bars = axs[0, 0].bar(
    top_5_cities["city"], 
    top_5_cities["population_total"], 
    color=["blue", "gray", "brown", "olive", "blue"], 
    alpha=0.7
)

# Añadir texto sobre las barras
for index, bar in enumerate(bars):
    axs[0, 0].text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 100,
        str(top_5_cities.iloc[index]["population_total"]),
        ha="center",
        va="bottom"
    )

# Configuración del gráfico
axs[0, 0].set_title("Ciudades Más Pobladas de California")
axs[0, 0].set_xlabel("Ciudades", fontsize=10)
axs[0, 0].set_ylabel("Habitantes (miles)", fontsize=10)
axs[0, 0].set_ylim(0, 5000)

# --------------- Segundo gráfico ---------------

axs[0, 1].hist(df["area_total_km2"], bins=50, color="steelblue", alpha=0.7, label="Hist. Área")

axs[0, 1].legend(fontsize=8, loc="upper right")
axs[0, 1].set_xlabel("Superficie (km$^2$)", fontsize=14, fontname="Bradley Hand ITC")
axs[0, 1].set_ylabel("Número de ciudades", fontsize=14, fontname="Bradley Hand ITC")

# --------------- Tercer gráfico ---------------

axs[1, 0].scatter(df["area_water_km2"], df["population_total"] / 1000, c="green", alpha=0.6)

axs[1, 0].set_title("Población por km$^2$ de agua")
axs[1, 0].set_xlabel("Área de agua en la ciudad (km$^2$)", fontsize= 10)
axs[1, 0].set_ylabel("Población(miles de habitantes)", fontsize= 10)

axs[1, 0].annotate(
    "Outliers",
    xy=(max_area_agua["area_water_km2"].iloc[0], max_area_agua["population_total"].iloc[0] / 1000),
    xytext=(60, 3500),
    arrowprops=dict(facecolor="black", arrowstyle="->"),
    fontsize=10,
    color="black",
)

axs[1, 0].annotate(
    "Outliers",
    xy=(max_poblacion["area_water_km2"].iloc[0], max_poblacion["population_total"].iloc[0] / 1000),
    xytext=(60, 3500),
    arrowprops=dict(facecolor="black", arrowstyle="->"),
    fontsize=10,
    color="black",
)

# --------------- Cuarto gráfico ---------------

axs[1, 1].boxplot(df["elevation_m"], vert= False, patch_artist=True, boxprops=dict(facecolor="red", color="black"), medianprops=dict(color="black"))

axs[1, 1].set_title("Diagrama de caja de la altitud de las ciudades de California", fontsize= 10)
axs[1, 1].set_xlabel("Altitud (m)")

plt.tight_layout()
plt.show()
```
