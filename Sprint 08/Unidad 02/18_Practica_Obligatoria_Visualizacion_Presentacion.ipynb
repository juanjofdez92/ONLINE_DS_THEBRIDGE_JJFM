### #1.2

Crea una funcin similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una funcin para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numrica del dataframe
3. Una columna numrica del dataframe diferente a la anterior.
4. Un columna categrica (Opcional, valor por defecto None)
5. Un valor para el tamao de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamao
6. Un sexto argumento "escala", que permita manipular la escala del tamao en el caso de que el quinto argumento sea una cadena. Este argumento multiplicar a los valores de la columna indicada en el parmetro anterior (as si quiero dividir por 10000, este sexto argumento tendr que valer 1/10000)

La funcin debe pintar el diagrama de dispersin de la columna numrica 1, en el eje X, frente a la columna numrica 2, en el eje Y, usando un cdigo de colores en funcin de la columna categrica y con el tamao de los puntos segn el argumento correspondiente. 

Si no se diera una columna categrica como argumento, el color no debe pasarse como parmetro a la funcin de Seaborn o Matplotlib que uses.

EXTRA: Aade un argumento que controle si la funcin pinta una leyenda para colores y tamaos o un colorbar y una leyenda para los tamaos. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesin en vivo)

```python
def pintar_dispersion(dataframe, col_x, col_y, col_cat=None, size_or_col=50, escala=1):
  
    for col in [col_x, col_y, col_cat if col_cat else None]:
        if col and col not in dataframe.columns:
            raise ValueError(f"La columna {col} no est en el dataframe.")

    if isinstance(size_or_col, str):
        if size_or_col not in dataframe.columns:
            raise ValueError(f"La columna {size_or_col} no est en el dataframe.")
        sizes = dataframe[size_or_col] * escala
    else:
        sizes = size_or_col

    plt.figure(figsize=(10, 6))

    if col_cat:
        sns.scatterplot(
            data=dataframe,
            x=col_x,
            y=col_y,
            hue=col_cat,
            size=sizes,
            sizes=(20, 200),
            legend="brief"
        )
    else:
        sns.scatterplot(
            data=dataframe,
            x=col_x,
            y=col_y,
            size=sizes,
            sizes=(20, 200),
            legend="brief"
        )

    plt.title("Diagrama de Dispersin")
    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.show()
```

```python
df = pd.read_csv("./data/titanic.csv")
df
```

**1. El Titanic fue un accidente con una tasa de mortalidad alta**

```python
survival_counts = df["survived"].value_counts(normalize=True) * 100

ax = sns.barplot(x=survival_counts.index, y=survival_counts.values)
plt.xticks([0, 1], ['Fallecidos', 'Supervivientes'])
plt.ylim(0, 100)

for i, percentage in enumerate(survival_counts.values):
    ax.text(i, percentage + 1, f'{percentage:.1f}%', ha='center', fontsize=12, color='black')

plt.title('Tasa de mortalidad en el Titanic')
plt.show()
```

**2. En esa ocasin los datos apoyan que se dio el "Las mujeres y los nios primero", pero...**


```python
df["group"] = df.apply(
    lambda row: "Mujeres y nios" if row["sex"] == "female" or row["age"] < 18 else "Hombres", axis=1
)

survival_rates = df.groupby("group")["survived"].mean().reset_index()
survival_rates["Tasa de supervivencia (%)"] = survival_rates["survived"] * 100

plt.figure(figsize=(8, 6))
ax = sns.barplot(data=survival_rates, x="group", y="Tasa de supervivencia (%)", hue="group", palette="Set2")

plt.title("Tasa de supervivencia por grupo", fontsize=16)
plt.xlabel("Grupo", fontsize=12)
plt.ylabel("Tasa de supervivencia (%)", fontsize=12)
plt.ylim(0, 100)

for p in ax.patches:
    ax.annotate(
        f'{p.get_height():.1f}%',  # Formato del porcentaje
        (p.get_x() + p.get_width() / 2., p.get_height()),  # Posicin
        ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
        textcoords='offset points'
    )

plt.show()
```

**3. Que la clase en la que se viaj tuvo una influencia significativa y podra explicar...**

```python
survival_by_class = df.groupby("pclass")["survived"].mean().reset_index()

plt.figure(figsize=(10, 6))
ax = sns.barplot(data=survival_by_class, x="pclass", y="survived", hue="pclass", palette="Set2")

plt.title("Tasa de supervivencia por clase", fontsize=16)
plt.ylabel("Tasa de supervivencia", fontsize=12)
plt.xlabel("Clase", fontsize=12)
plt.ylim(0, 1)

for i, row in survival_by_class.iterrows():
    ax.text(i, row["survived"] + 0.02, f'{row["survived"]:.2f}', ha='center', fontsize=11, color='black')

plt.show()
```

**4. Las diferencias de tasas de supervivencia entre mujeres y nios.**  


```python
df["is_child"] = df["age"] < 18

women_and_children = df[(df["sex"] == "female") | (df["is_child"] == True)]

survival_rates = pd.DataFrame({
    "Grupo": ["Mujeres", "Nios"],
    "Tasa de supervivencia": [
        women_and_children[women_and_children["sex"] == "female"]["survived"].mean(),
        women_and_children[women_and_children["is_child"]]["survived"].mean()
    ]
})

plt.figure(figsize=(8, 6))
ax = sns.barplot(
    data=survival_rates, 
    x="Grupo", 
    y="Tasa de supervivencia",
    hue="Grupo", 
    palette="Set2"
)

plt.title("Tasa de supervivencia entre mujeres y nios", fontsize=16)
plt.xlabel("Grupo", fontsize=12)
plt.ylabel("Tasa de supervivencia", fontsize=12)
plt.ylim(0, 1)

# Aadir los porcentajes sobre las barras
for i, row in survival_rates.iterrows():
    ax.text(
        i, 
        row["Tasa de supervivencia"] + 0.02, 
        f'{row["Tasa de supervivencia"]:.2f}', 
        ha='center', 
        fontsize=11
    )

plt.show()
```

**5. Adems aunque el puerto de embarque parece influir, se podra asumir que lo fue el tipo de pasaje que embarc en cada uno.**

```python
survival_by_city = df.groupby("embark_town")["survived"].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=survival_by_city, x="embark_town", y="survived", hue="embark_town", palette="coolwarm")

plt.title("Tasa de supervivencia por ciudad de embarque", fontsize=16)
plt.xlabel("Ciudad de embarque", fontsize=12)
plt.ylabel("Tasa de supervivencia", fontsize=12)
plt.ylim(0, 1)

for i, row in survival_by_city.iterrows():
    plt.text(i, row["survived"] + 0.02, f'{row["survived"]:.2f}', ha='center', fontsize=11)

plt.show()
```

```python
class_distribution = df.groupby(["embark_town", "pclass"])["pclass"].count().unstack(fill_value=0)
class_distribution_percentage = (class_distribution.div(class_distribution.sum(axis=1), axis=0) * 100).reset_index()

class_distribution_percentage.columns = ["Puerto", "Primera Clase", "Segunda Clase", "Tercera Clase"]

data = class_distribution_percentage.melt(id_vars="Puerto", var_name="Clase", value_name="Porcentaje")

plt.figure(figsize=(10, 6))
ax = sns.barplot(data=data, x="Puerto", y="Porcentaje", hue="Clase", palette="viridis")

# Aadir los porcentajes encima de las barras
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='baseline', 
                fontsize=10, color='black', xytext=(0, 5), 
                textcoords='offset points')

plt.title("Porcentaje de pasajeros por clase y puerto de embarque", fontsize=16)
plt.xlabel("Puerto de embarque", fontsize=12)
plt.ylabel("Porcentaje (%)", fontsize=12)
plt.legend()
plt.show()
```
